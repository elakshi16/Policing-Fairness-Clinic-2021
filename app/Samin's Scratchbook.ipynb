```python
import numpy as np
import os

impG = graph.Graph()
impG.createNodes(from_file = "./out/nodes.pkl")

x = []
y = []
scalval = []
for key, node in myG.nodes.items():
    x.append(key[0])
    y.append(key[1])
    scalval.append(node.precinct_ID)
# size and color:

# plot
fig, ax = plt.subplots()

fig.set_size_inches(20,20)

# ax.scatter(x, y, s=arrCnt, alpha=0.2)
if not os.path.exists('./out/graphics'):
    os.makedirs('./out/graphics')
ax.scatter(x, y, s=1, c=scalval, cmap=cm.get_cmap('tab20')) # color instead of size 

ax.set(xlim=(0, 350), xticks=np.arange(1, 8),
       ylim=(0, 475), yticks=np.arange(1, 8))

plt.colorbar(ax=ax)

plt.savefig("./out/graphics/precinct_divisions.png")
```
